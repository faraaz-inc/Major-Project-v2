```python
import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier
import joblib

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
BASE_PATH = "/content/drive/MyDrive"

TRAIN_PATH = f"{BASE_PATH}/MajorProject/CIC_IoT_2023/Full_Dataset/train_working.csv"
TEST_PATH = f"{BASE_PATH}/MajorProject/CIC_IoT_2023/Full_Dataset/test_frozen.csv"

MODEL_A_SAVE_PATH = f"{BASE_PATH}/MajorProject/IDS_ModelA_v3/modelA_LGBM.joblib"
MODEL_B_SAVE_PATH = f"{BASE_PATH}/MajorProject/IDS_ModelB_v3/modelB_LGBM.joblib"

RANDOM_SEED = 42
```

```python
print("Loading Training data")
train_df = pd.read_csv(TRAIN_PATH, low_memory=False)

print("Loading Test data")
test_df = pd.read_csv(TEST_PATH, low_memory=False)

print("Datasets Loaded Succesfully")
```

```python
print(f"Shape of training data: {train_df.shape}")
print(f"Shape of test data: {test_df.shape}")
```

```python
LABEL_COL = 'label'

train_df = train_df.dropna(subset=[LABEL_COL])
test_df = test_df.dropna(subset=[LABEL_COL])

print("Train shape after cleaning", train_df.shape)
print("Test shape after cleaning", test_df.shape)
```

### Split features and target

```python
x_train = train_df.drop(columns=[LABEL_COL])
y_train = train_df[LABEL_COL]

x_test = test_df.drop(columns=[LABEL_COL])
y_test = test_df[LABEL_COL]

del train_df, test_df
```

```python
print("NANs in x_train:", x_train.isna().sum().sum())
print("NaNs in X_test:", x_test.isna().sum().sum())
print("NaNs in y_train:", y_train.isna().sum())
print("NaNs in y_test:", y_test.isna().sum())
```

### Encode labels

```python
le = LabelEncoder()

y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

class_names = le.classes_

print("Number of classes", len(class_names))
```

### LightGBM Model

```python
lgbm_model = LGBMClassifier(
    n_estimators=400,          # more trees → better performance
    learning_rate=0.05,        # lower LR → better generalization
    num_leaves=128,            # higher complexity
    max_depth=-1,              # no depth limit
    min_child_samples=20,      
    subsample=0.8,             # bagging
    colsample_bytree=0.8,      # feature sampling
    reg_alpha=0.1,             # L1 regularization
    reg_lambda=1.0,            # L2 regularization
    n_jobs=-1,
    random_state=RANDOM_SEED,
    verbose=1
)
```

```python
print("Training LightGBM on", x_train.shape[0], "rows")

lgbm_model.fit(x_train, y_train_enc)
```
